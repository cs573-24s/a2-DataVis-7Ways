```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV file
df = pd.read_csv('penglings.csv')
```

```python
# Get the unique species
unique_species = df["species"].unique()
print(unique_species)
```

```python
# Adjusting the code for colors, legend, and bubble sizes
plt.figure(figsize=(10, 6))

# Color mapping
species_colors = {"Adelie": "darkorange", "Gentoo": "darkgreen", "Chinstrap": "purple"}

# Normalizing and scaling bill lengths to use as circle sizes based on the bill lengths
min_bill_length = df["bill_length_mm"].min()
max_bill_length = df["bill_length_mm"].max()
normalized_bill_length = (df["bill_length_mm"] - min_bill_length) / (max_bill_length - min_bill_length)
scaled_sizes = normalized_bill_length * 5 + 1  # Decreased scaling factor

for species in df["species"].unique():
    species_data = df[df["species"] == species]
    sizes = scaled_sizes[species_data.index] * 50  # Adjusted size calculation

    # Plotting each species with a label for the legend
    plt.scatter(
        species_data["flipper_length_mm"], 
        species_data["body_mass_g"], 
        s=sizes, 
        alpha=0.7, 
        color=species_colors[species], 
        edgecolors=species_colors[species], 
        linewidths=1.5, 
        label=species
    )

# Adding labels and title
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Body Mass (g)")
plt.title("Scatter Plot of Penguins: Flipper Length vs Body Mass")

# Adding legend
plt.legend(title="Species", loc="upper right", bbox_to_anchor=(1.2, 1))
```
