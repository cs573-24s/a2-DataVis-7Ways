```python
import pandas as pd
from plotnine import ggplot, aes, labs, geom_point, labs, scale_color_manual, facet_wrap
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('penglings.csv')
df
```

```python
# remove rows that contaion null value in 'bill_length_mm' column 
df = df.dropna(axis=0, subset='bill_length_mm')
```

```python
(ggplot(df,
        aes(x='flipper_length_mm', 
            y='body_mass_g',
               size='bill_length_mm',
               color='species'))
        + geom_point(alpha=.7)
        + labs(x='Flipper Length (mm)',
               y='Body Mass (g)',
               color='Species',
               size='Bill Length (mm)',
               title='Python + Plotnine (ggplot)')
        + scale_color_manual(values=['darkorange', 'darkorchid', 'cadetblue'])
)
```

```python
(ggplot(df,
        aes(x='flipper_length_mm', 
            y='body_mass_g',
               size='bill_length_mm',
               color='species'))
        + geom_point(alpha=.7)
        + facet_wrap('island', nrow=3)
        + scale_color_manual(values=['darkorange', 'darkorchid', 'cadetblue'])
)
```
