```python
import matplotlib.pyplot as plt
import pandas as pd


penglings = pd.read_csv('penglings.csv')


adelie = penglings[penglings['species'] == 'Adelie']
gentoo = penglings[penglings['species'] == 'Gentoo']
chinstrap = penglings[penglings['species'] == 'Chinstrap']


colors = {'Adelie': '#FF5733', 'Gentoo': '#00FF00', 'Chinstrap': '#800080'}
size_scale = 100 / penglings['bill_length_mm'].max()  # Scale factor for sizes

plt.figure(figsize=(10, 6))
plt.scatter(adelie['flipper_length_mm'], adelie['body_mass_g'], 
            s=adelie['bill_length_mm'] * size_scale, c=colors['Adelie'], label='Adelie', alpha=0.8)
plt.scatter(gentoo['flipper_length_mm'], gentoo['body_mass_g'], 
            s=gentoo['bill_length_mm'] * size_scale, c=colors['Gentoo'], label='Gentoo', alpha=0.8)
plt.scatter(chinstrap['flipper_length_mm'], chinstrap['body_mass_g'], 
            s=chinstrap['bill_length_mm'] * size_scale, c=colors['Chinstrap'], label='Chinstrap', alpha=0.8)


plt.title('Penguins ScatterPlot')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')


plt.xlim(160, 230)
plt.ylim(2000, 7000)


plt.text(230, 7000, 'Small Ball (bill length mm: 40)', color='black', ha='right')
plt.text(230, 6800, 'Large Ball (bill length mm: 50)', color='black', ha='right')


plt.legend()
plt.show()
```
