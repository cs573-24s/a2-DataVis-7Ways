```python
import altair as alt
import pandas as pd

# reading the csv
data = pd.read_csv('./penglings.csv')

# preform color matching to the specific species
color_mapping = alt.Scale(domain=['Adelie', 'Chinstrap', 'Gentoo'],
                          range=['orange', 'purple', 'green'])  

# create the chart using Chart
alt.Chart(data, title='Python + Altair Viz').mark_point(filled=True).encode(
    x=alt.X('flipper_length_mm', scale=alt.Scale(zero=False)), # set the x and y to not start at 0
    y=alt.Y('body_mass_g', scale=alt.Scale(zero=False)),
    color=alt.Color('species', scale=color_mapping), # map to the correct color
    size='bill_length_mm:Q', # change the size of each data point depending on the bill length
    opacity=alt.value(0.7) # change the opacity to mimic the transparent effect like the example
)
```
