```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("../penglings.csv")
```

```python
fig, ax = plt.subplots(figsize = ( 6 , 5)) 
palette = ["#fba044","#479f9f","#9933cb"]
graph = sns.scatterplot(data=data, ax=ax, x="flipper_length_mm", y="body_mass_g", hue="species", size="bill_length_mm", alpha=0.7, palette=palette)
graph.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
ax.set_xlabel( "Flipper Length (mm)" , size = 12 )
ax.set_ylabel( "Body Mass (g)" , size = 12 )
```
