```python
import pandas as pd
import seaborn as sns
import matplotlib as plt

data = pd.read_csv("penglings.csv").dropna(subset=['body_mass_g', 'flipper_length_mm'])

def get_bill_size(bill_length_mm):
    return 40 if bill_length_mm < 45 else 50

data['bill_size'] = data['bill_length_mm'].apply(get_bill_size)
data['Body Mass (g)'] = data['body_mass_g']
sns.scatterplot(x="flipper_length_mm", y="Body Mass (g)", data=data, size='bill_size', alpha=0.8, hue='species').set_xlabel("Flipper Length (mm)")
```
