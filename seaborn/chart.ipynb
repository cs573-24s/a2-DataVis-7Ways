```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
data = pd.read_csv("../penglings.csv")
pd.read_csv("../penglings.csv")
```

```python
#https://stackoverflow.com/questions/14885895/color-a-scatter-plot-by-column-values
palette = ["#fba044","#479f9f","#9933cb"]
with plt.rc_context({'legend.labelspacing': 1.2}):
    pengins = sns.relplot(
        data=data, 
        x="flipper_length_mm", 
        y="body_mass_g", 
        alpha=0.6, 
        hue='species', 
        size='bill_depth_mm', 
        palette=palette
        ).set(
            xlabel="Flipper Length (mm)", 
            ylabel="Body Mass (g)"
            )
    plt.show()
```
