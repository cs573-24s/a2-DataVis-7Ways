```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load dataset
df = pd.read_csv("./penglings.csv", index_col=0)
df.head()
```

```python
# Create a color palette for the species
palette = ['#FF8C00', '#008B8B', '#9932CC']
sns.set_palette(palette)
```

```python
plt.figure(figsize=(20, 15))
sns.scatterplot(
    data=df,
    x='flipper_length_mm',
    y='body_mass_g',
    hue='species',
    size='bill_length_mm',
    sizes=(50, 500),
    alpha=0.8
)

# Set the labels
plt.title("Penguin Dataset")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Body Mass (g)")
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.5)
plt.show()
```

```python
# save the plot
plt.savefig("seaborn.png")
```
