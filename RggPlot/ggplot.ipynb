```r


library(ggplot2)
library(readr) # For read_csv


penglings <- read_csv("penglings.csv")


ggplot(penglings, aes(x = flipper_length_mm, y = body_mass_g, color = species, size = bill_length_mm)) +
  geom_point(alpha = 0.8) + 
  scale_color_manual(values = c("Adelie" = "#FF5733", "Gentoo" = "#00FF00", "Chinstrap" = "#800080")) +
  scale_size_continuous(range = c(4, 9)) + 
  theme_minimal() +
  labs(x = "Flipper Length (mm)", y = "Body Mass (g)", size = "Bill Length (mm)") +
  theme(legend.position = "right") + 
  guides(color = guide_legend(title = "Species"), 
         size = guide_legend(override.aes = list(alpha = 1))) 


```
