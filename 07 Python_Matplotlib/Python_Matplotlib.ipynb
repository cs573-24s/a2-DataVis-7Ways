```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')

# Load the dataset
# You need to change the path if the csv file is in different path
data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/data visualization a2/penglings.csv')
```

```python
# Drop rows with missing values in the relevant columns
data = data.dropna(subset=['flipper_length_mm', 'body_mass_g', 'bill_length_mm'])

# Create a color mapping for the species
species_colors = {
    'Adelie': 'orange',
    'Chinstrap': 'purple',
    'Gentoo': 'green'
}

# Map the species to colors
data['color'] = data['species'].map(species_colors)

# Plot the scatter chart
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    data['flipper_length_mm'],
    data['body_mass_g'],
    alpha=0.8,
    c=data['color'],
    s=data['bill_length_mm'] * 5,  # Multiply by a factor to scale up the sizes for visibility
)

# Label the axes
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')

# Set the scale for the axes
plt.xlim(data['flipper_length_mm'].min() - 10, data['flipper_length_mm'].max() + 10)
plt.ylim(data['body_mass_g'].min() - 500, data['body_mass_g'].max() + 500)

# Add ticks at reasonable intervals
plt.xticks(range(int(data['flipper_length_mm'].min()), int(data['flipper_length_mm'].max()), 10))
plt.yticks(range(int(data['body_mass_g'].min()), int(data['body_mass_g'].max()), 500))
```

```python
# Show the plot
plt.show()
```
