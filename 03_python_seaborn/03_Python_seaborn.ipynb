```python
import pandas as pd
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('penglings.csv')
```

```python
df
```

```python
sns.set_style("darkgrid")
sns.set_context("paper")
```

```python
fig = sns.relplot(data=df,
                  x="flipper_length_mm", y="body_mass_g", 
                  hue="species", 
                  size="bill_length_mm", sizes=(40, 500), 
                  alpha=.7, palette="muted",
                  aspect = 1.5).set(xlabel='Flipper Length (mm)',
                                    ylabel='Body Mass (g)',
                                    title='Python + Seaborn');
fig.savefig('python_seaborn.png')
```
